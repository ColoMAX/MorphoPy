## Loading in a morphology

MorphoPy can read `swc` files in their standard format.

    id x y z type radius pid
   

When loading in a file the data is checked and a short summary is returned. **Please note**, that some functionalities only work when a complete neuron is passed (having a defined soma, axon and dendrites).

```python
m = morphopy.Morph(data='data/C4.swc')
```

## Morphometric summary statistics

MorphoPy can calculate morphometric statistics about a morphology. 
    

```python
m.processing()
```

```python
m.show_summary()
```
